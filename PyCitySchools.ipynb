```python
# Add dependencies.
import os
import pandas as pd
```

```python
#Files to load
school_data_to_load = os.path.join(".","resources","schools_complete.csv")
student_data_to_load = os.path.join(".","resources","students_complete.csv")

print(school_data_to_load)
print(student_data_to_load)
```

```python
# Read the school data file and store it in a Pandas DataFrame.
df_school_data = pd.read_csv(school_data_to_load)
df_school_data
```

```python
# Read the student data file and store it in a Pandas DataFrame.
df_student_data = pd.read_csv(student_data_to_load)
df_student_data.head(10)
```

```python
# Add each prefix and suffix to remove to a list.
list_prefixes_suffixes = ['Dr. ','Miss ','Mr. ','Mrs. ','Ms. ',' DDS',' DVM',' MD',' PhD']
```

```python
# Iterate through the "prefixes_suffixes" list
# and replace them with an empty space, ""
#when it appears in the student's name.
for affix in list_prefixes_suffixes:
    df_student_data["student_name"] = df_student_data["student_name"].str.replace(affix,"")

df_student_data.head(10)
```

```python
# Combine the data into a single dataset.
df_school_data_complete = pd.merge(df_student_data, df_school_data, on=['school_name','school_name'])

df_school_data_complete.head(10)
```

```python
# Get the total number of students.

# count_students = df_school_data_complete.count()
count_students = df_school_data_complete['Student ID'].count()

count_students
```

```python
# Calculate the total number of schools.
count_schools = df_school_data['school_name'].count()

count_schools
```

```python
# Calculate the total number of schools (from the combined data)
count_schools = len(df_school_data_complete['school_name'].unique())

count_schools
```

```python
# Calculate the total budget.
total_budget = df_school_data['budget'].sum()

total_budget
```

```python
# Calculate the average reading score.
average_reading_score = df_school_data_complete['reading_score'].mean()

average_reading_score
```

```python
# Calculate the average math score.
average_math_score = df_school_data_complete['math_score'].mean()

average_math_score
```

```python
series_passing_math = 70 <= df_school_data_complete['math_score']
series_passing_reading = 70 <= df_school_data_complete['reading_score']

series_passing_reading
```

```python
# Get all the students who are passing math.
df_passing_math = df_school_data_complete[70 <= df_school_data_complete['math_score']]

df_passing_math.head(10)
```

```python
# Get all the students who are passing reading.
df_passing_reading = df_school_data_complete[70 <= df_school_data_complete['reading_score']]

df_passing_reading.head(10)
```

```python
# Calculate the number of students passing math.
count_passing_math = df_passing_math['Student ID'].count()

# Calculate the number of students passing reading.
count_passing_reading = df_passing_reading['Student ID'].count()

print(count_passing_math)
print(count_passing_reading)
```

```python
# Calculate the percentage of students who passed math.
pct_passing_math = count_passing_math * 100 / float(count_students)

# Calculate the percentage of students who passed reading.
pct_passing_reading = count_passing_reading * 100 / float(count_students)

print(pct_passing_math)
print(pct_passing_reading)
```

```python
# Calculate the students who passed both math and reading.
df_passing_math_reading = df_school_data_complete[
    (70 <= df_school_data_complete['math_score'])
    &
    (70 <= df_school_data_complete['reading_score'])
]

df_passing_math_reading.head(10)
```

```python
# Calculate the number of students who passed both math and reading.
count_passing_math_reading = df_passing_math_reading['Student ID'].count()

count_passing_math_reading
```

```python
# Calculate the overall passing percentage.
pct_passing_math_reading = count_passing_math_reading * 100 / float(count_students)

pct_passing_math_reading
```

```python
# Adding a list of values with keys to create a new DataFrame.
df_district_summary = pd.DataFrame([{
    'Total Schools':count_schools,
    'Total Students':count_students,
    'Total Budget':total_budget,
    'Average Math Score':average_math_score,
    'Average Reading Score':average_reading_score,
    '% Passing Math':pct_passing_math,
    '% Passing Reading':pct_passing_reading,
    '% Passing Overall':pct_passing_math_reading
}])

df_district_summary
```

```python
# Define a function that converts counts into percentages
def pct(num,denom):
    return num * 100 / float(denom)
```

```python
pct(count_passing_math,count_students)
```

```python
# Format the "Total Students" to have the comma for a thousands separator.
df_district_summary['Total Students'] = df_district_summary['Total Students'].map("{:,}".format)

df_district_summary['Total Students']
```

```python
# Format "Total Budget" to have the comma for a thousands separator, a decimal separator, and a "$".
df_district_summary['Total Budget'] = df_district_summary['Total Budget'].map("${:,.2f}".format)

df_district_summary['Total Budget']
```

```python
# Format the columns.
df_district_summary['Average Math Score'] = df_district_summary['Average Math Score'].map("{:.1f}".format)
df_district_summary['Average Reading Score'] = df_district_summary['Average Reading Score'].map("{:.1f}".format)
df_district_summary['% Passing Math'] = df_district_summary['% Passing Math'].map("{:.0f}".format)
df_district_summary['% Passing Reading'] = df_district_summary['% Passing Reading'].map("{:.0f}".format)
df_district_summary['% Passing Overall'] = df_district_summary['% Passing Overall'].map("{:.0f}".format)

df_district_summary
```
